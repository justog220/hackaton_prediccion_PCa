# 02. Cleaning

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/interim/data_full.csv", index_col="sample_id")

df
```

## Missing Values

Droping `grado_clinico` missing values.

```python
missing_counts = df.isnull().sum()
print(missing_counts[missing_counts > 0])
```

```python
df_without_na = df.dropna(subset=['grado_histologico']).copy()

df_without_na.to_csv("../data/interim/cleaning/data_no_na.csv", index=True)
```

## Scale Features

Following [Sklearn preprocessing guide](https://scikit-learn.org/stable/modules/preprocessing.html)

```python

X = df.iloc[:, :-3]  # gene expression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df_scaled = pd.DataFrame(X_scaled, index=df.index, columns=X.columns)
df_scaled[['cohorte','relapse','grado_histologico']] = df[['cohorte','relapse','grado_histologico']]
df_scaled.to_csv("../data/interim/cleaning/data_scaled.csv", index=True)
```

## Dataset with Scaling and no Na

```python
X = df_without_na.iloc[:, :-3]  # gene expression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

df_scaled = pd.DataFrame(X_scaled, index=df_without_na.index, columns=X.columns)
df_scaled[['cohorte','relapse','grado_histologico']] = df_without_na[['cohorte','relapse','grado_histologico']]
df_scaled.to_csv("../data/interim/cleaning/data_without_na_scaled.csv", index=True)
```
